```
#dy/dx=10-y 空気抵抗のある物体の落下速度の方程式 4次Runge-Kutta法
import numpy as np
import matplotlib.pyplot as plt
 
X = 50
n = 1000
h = X / n
x = np.arange(0,X,h)

print(x)
 
def f(x,y):
    return 10-y

Y_0 = 1
 
Y = np.empty(n)
Y[0] = Y_0

for i in range(n-1):
    k_1 = f(x[i],Y[i])
    k_2 = f(x[i]+h/2,Y[i]+k_1 / 2)
    k_3 = f(x[i]+h/2,Y[i] + k_2 /2)
    k_4 = f(x[i]+h,Y[i] + k_3)        
    Y[i+1] = Y[i] + h/6 * (k_1 + 2*k_2 + 2*k_3 + k_4 )
 
plt.plot(x,Y, label="RungeKutta4")

print(Y)

print(Y[2])
 
plt.legend()
plt.show()
```

```
#2次Runge-Kutta法
import numpy as np
import matplotlib.pyplot as plt
 
X = 50
n = 1000
h = X / n
x = np.arange(0,X,h)

print(x)
 
def f(x,y):
    return 10-y

Y_0 = 1
 
Y = np.empty(n)
Y[0] = Y_0

for i in range(n-1):
    k_1 = f(x[i],Y[i])
    k_2 = f(x[i]+h,Y[i]+h*k_1)      
    Y[i+1] = Y[i] + h/2 * (k_1 + k_2)
 
plt.plot(x,Y, label="RungeKutta2")

print(Y)

print(Y[2])
 
plt.legend()
plt.show()
```

```
#Euler法
import numpy as np
import matplotlib.pyplot as plt
 
X = 50
n = 1000
h = X / n
x = np.arange(0,X,h)

print(x)
 
def f(x,y):
    return 10-y

Y_0 = 1
 
Y = np.empty(n)
Y[0] = Y_0

for i in range(n-1):     
    Y[i+1] = Y[i] + h * f(x[i],Y[i])
 
plt.plot(x,Y, label="Euler")

print(Y)

print(Y[2])
 
plt.legend()
plt.show()
```
