```
import numpy as np
import matplotlib.pyplot as plt
 
X = 300
n = 100000
h = X / n
x = np.arange(0,X,h)

def f(x,y,z):
    return -beta*y*z

def g(x,y,z):
    return beta*y*z-gamma*z

N = 1000000
beta=0.2/N
gamma=0.07

Y_0 = N-10
Z_0 = 10
W_0 = 0

Y = np.empty(n)
Z = np.empty(n)
W = np.empty(n)
Y[0] = Y_0
Z[0] = Z_0
W[0] = W_0
 
for i in range(n-1):
    k_1 = f(x[i],Y[i],Z[i])
    l_1 = g(x[i],Y[i],Z[i])
    k_2 = f(x[i]+h/2, Y[i] +h/2 * k_1 , Z[i]+h/2*l_1)
    l_2 = g(x[i]+h/2, Y[i]+ h/2 * k_1 , Z[i]+h/2*l_1)
    k_3 = f(x[i]+h/2, Y[i] +h/2 * k_2 , Z[i]+h/2*l_2)
    l_3 = g(x[i]+h/2, Y[i] +h/2 * k_2 , Z[i]+h/2*l_2)
    k_4 = f(x[i]+h, Y[i] +h * k_3 , Z[i]+h*l_3)
    l_4 = g(x[i]+h, Y[i] +h * k_3 , Z[i]+h*l_3)
    Y[i+1] = Y[i] + h/6 * (k_1 + 2*k_2 + 2*k_3 + k_4 )
    Z[i+1] = Z[i] + h/6 * (l_1 + 2*l_2 + 2*l_3 + l_4 )
    W[i+1] = N-Y[i+1]-Z[i+1]
 
plt.plot(x,Y, label="Susceptible")
plt.plot(x,Z, label="Infected")
plt.plot(x,W, label="Removed")

 
plt.legend()
plt.show()
```
