```
#dy/dx=2y-yz dz/dx=yz-z Lotka-Volterra(ロトカ・ヴォルテラ)方程式　x:時刻　y(x);時刻xにおける被食者の数　z(x):時刻xにおける捕食者の数 4次Runge-Kutta法


import numpy as np
import matplotlib.pyplot as plt
 
X = 50
n = 100000
h = X / n
x = np.arange(0,X,h)

def f(x,y,z):
    return 2*y-y*z

def g(x,y,z):
    return y*z-z

Y_0 = 10
Z_0 = 2

Y = np.empty(n)
Z = np.empty(n)
Y[0] = Y_0
Z[0] = Z_0
 
for i in range(n-1):
    k_1 = f(x[i],Y[i],Z[i])
    l_1 = g(x[i],Y[i],Z[i])
    k_2 = f(x[i]+h/2, Y[i] +h/2 * k_1 , Z[i]+h/2*l_1)
    l_2 = g(x[i]+h/2, Y[i]+ h/2 * k_1 , Z[i]+h/2*l_1)
    k_3 = f(x[i]+h/2, Y[i] +h/2 * k_2 , Z[i]+h/2*l_2)
    l_3 = g(x[i]+h/2, Y[i] +h/2 * k_2 , Z[i]+h/2*l_2)
    k_4 = f(x[i]+h, Y[i] +h * k_3 , Z[i]+h*l_3)
    l_4 = g(x[i]+h, Y[i] +h * k_3 , Z[i]+h*l_3)
    Y[i+1] = Y[i] + h/6 * (k_1 + 2*k_2 + 2*k_3 + k_4 )
    Z[i+1] = Z[i] + h/6 * (l_1 + 2*l_2 + 2*l_3 + l_4 )
 
plt.plot(x,Y, label="y")
plt.plot(x,Z, label="z")
 
plt.legend()
plt.show()
```

```
#2次Runge-Kutta法

import numpy as np
import matplotlib.pyplot as plt
 
X = 50
n = 100000
h = X / n
x = np.arange(0,X,h)

def f(x,y,z):
    return 2*y-y*z

def g(x,y,z):
    return y*z-z

Y_0 = 10
Z_0 = 2

Y = np.empty(n)
Z = np.empty(n)
Y[0] = Y_0
Z[0] = Z_0
 
for i in range(n-1):
    k_1 = f(x[i],Y[i],Z[i])
    l_1 = g(x[i],Y[i],Z[i])
    k_2 = f(x[i]+h, Y[i] +h * k_1 , Z[i]+h*l_1)
    l_2 = g(x[i]+h, Y[i]+ h * k_1 , Z[i]+h*l_1)
    Y[i+1] = Y[i] + h/2 * (k_1 + k_2)
    Z[i+1] = Z[i] + h/2 * (l_1 + l_2)
 
 
plt.plot(x,Y, label="y")
plt.plot(x,Z, label="z")
 
plt.legend()
plt.show()
```

```
#Euler法

import numpy as np
import matplotlib.pyplot as plt
 
X = 50
n = 100000
h = X / n
x = np.arange(0,X,h)

def f(x,y,z):
    return 2*y-y*z

def g(x,y,z):
    return y*z-z

Y_0 = 10
Z_0 = 2

Y = np.empty(n)
Z = np.empty(n)
Y[0] = Y_0
Z[0] = Z_0
 
for i in range(n-1):
    Y[i+1] = Y[i] + h* f(x[i],Y[i],Z[i])
    Z[i+1] = Z[i] + h* g(x[i],Y[i],Z[i])
 
 
plt.plot(x,Y, label="y")
plt.plot(x,Z, label="z")
 
plt.legend()
plt.show()
```
